```python
import matplotlib.pyplot as plt
from PIL import Image

def display_images_side_by_side(image_path, size=(512, 512)):
    # Load the image from the given path and ensure it's in RGB mode.
    original_img = Image.open(image_path).convert('RGB')
    
    # Resize the image using the computed dimensions.
    resized_img = original_img.resize(size)
    
    # Create a figure with two subplots (side by side)
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
    
    # Plot the original image.
    ax1.imshow(original_img)
    ax1.set_title("Original Image")
    ax1.axis('off')  # Hide axes ticks
    
    # Plot the resized image.
    ax2.imshow(resized_img)
    ax2.set_title("Resized Image")
    ax2.axis('off')  # Hide axes ticks
    
    # Display the images.
    plt.tight_layout()
    plt.show()

# Example usage:
if __name__ == '__main__':
    image_path = '/Users/leonardo/Pictures/fish_image_sample/fam_gen_sp/Cirripectes variolosus.jpeg'  # Replace with your image file path.
    display_images_side_by_side(image_path, size=(128, 128))
```
